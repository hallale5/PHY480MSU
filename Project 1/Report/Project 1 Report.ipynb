Introduction:

This lab project is based around testing the relative error due to rounding in a computational program. The time for the process to complete was also tracked, as it was ran as a brute force method as well as an LU Decomposition. The time between the two was then compared to see the differences between the two. 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

'''Here we are importing the data that was computed using the C++ Program for the brute force method'''

a = np.loadtxt('out1',skiprows=2) #skip rows makes it only pull in the numbers needed, leaving out the labels row and time row
b = np.loadtxt('out2',skiprows=2)
c = np.loadtxt('out3',skiprows=2)
d = np.loadtxt('out4',skiprows=2)
e = np.loadtxt('out5',skiprows=2)
f = np.loadtxt('out6',skiprows=2)
```

```python

plt.figure(figsize=(18,6))
plt.plot (a[:,0],a[:,3],'b',b[:,0],b[:,3],'r',c[:,0],c[:,3],'g',d[:,0],d[:,3],'y',e[:,0],e[:,3],'teal',f[:,0],f[:,3],'orange')
plt.title ('Brute Force, data points vs Log(error)')
plt.xlabel('x data point')
plt.ylabel('log(error)')
labels=('10','100','1000','10^4','10^5','10^6')
plt.legend(labels,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

the above is for values of 10^1 through 10^6

```python
g = np.loadtxt('out7',skiprows=2)
```

```python
plt.figure(figsize=(18,6))
plt.plot(g[:,0],g[:,3],'b',linewidth=.05)
plt.title ('Plot of 10^7x10^7 matrix with LU Decomposition')
plt.xlabel('x data point')
plt.ylabel('log(error)')
```

Above is for using the LU decomp with a value of 10^7, as can be seen the error with this attempt at high precision is unpredictable compared to 10^6 and below. With the lower amount of data points we see a consistant error, which is a straight line whereas the 10^7 is all over the place, more precise than some of the others but yet not reliable. THis had to be done with the LU decomposition because the program would crash with the other method. 10^7 also appears to be the maximum that the LU can handle as attempting to go to 10^8 crashed the program.
